```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle
import pandas as pd

df = pd.read_csv('./raw_data/stats.tsv', sep='\t', index_col=0)
# print(df)

with open('trees/con_50_majrule.newick', 'r') as in_f:
    with open('trees/mb_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = '"'+ df['description'][el]+'"'
                    line = line.replace(el, new_name)
            out_f.write(line)
            print(line)
```

```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle
import pandas as pd

df = pd.read_csv('./raw_data/stats.tsv', sep='\t', index_col=0)
# print(df)

with open('trees/con_50_majrule.newick', 'r') as in_f:
    with open('trees/mb_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = df['description'][el].replace(',', '')
                    line = line.replace(el, new_name)
            out_f.write(line)
            print(line)
```
