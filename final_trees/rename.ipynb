```python
import pandas as pd
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle

pth = '/home/yulia/parampampam/'
```

#### SSU ML tree

```python
stats = 'final_trees/SSU_stats.tsv'

df = pd.read_csv(pth+stats, sep='\t', index_col=0)

with open(pth+'trees/18S/tree_6/ml/RAxML_bipartitions.result', 'r') as in_f:
    with open(pth+'final_trees/ml/SSU/RAxML_for_seaview.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = df['new'][el]
                    line = line.replace(el, new_name)
            out_f.write(line)

#   seaview            
            
with open(pth+'final_trees/ml/SSU/RAxML_rooted.newick', 'r') as in_f:
    with open(pth+'final_trees/ml/SSU/RAxML_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for ind, name in zip(df.index, df.new):
                name = name.lstrip(' ')
                if name in line:
                    new_name = '"' + df['new'][ind] + '"'
                    line = line.replace(name, new_name)
            out_f.write(line)
            
ours = list(df[df.current_study == True].new)
errs = list(df[df.erroneous == True].new)

t = Tree(pth+'final_trees/ml/SSU/RAxML_ete_tree.newick', quoted_node_names=True)

leafs = t.get_leaf_names()
leafs.remove('Paramecium nephridiatum dup BMS16-23 MK764889.1')
t.prune(leafs)

for node in t.traverse():
    node.img_style['size'] = 0

style1 = NodeStyle()
style1['bgcolor'] = 'LightBlue'

style2 = NodeStyle()
style2['bgcolor'] = 'LightCoral'

for leaf in t.iter_leaves():
    if str(leaf)[3:] in set(ours):
        leaf.img_style = style1
    if str(leaf)[3:] in set(errs):
        leaf.img_style = style2
    
t.ladderize(direction=1)
ts = TreeStyle()
ts.show_branch_support = True
ts.scale = 10000
t.show(tree_style=ts)
t.render(pth+'final_trees/ml/SSU/RAxML_ete_tree.pdf', dpi=1200, tree_style=ts)
```
