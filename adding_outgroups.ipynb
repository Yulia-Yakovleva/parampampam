```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle
import pandas as pd

df = pd.read_csv('./raw_data/stats/stats_final.tsv', sep='\t', index_col=0)
# print(df)

with open('trees/ml_tree_6_more_outgroup/RAxML_bipartitions.result', 'r') as in_f:
    with open('trees/ml_tree_6_more_outgroup/RAxML_for_seaview.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = df['description'][el].replace(',', '')
                    line = line.replace(el, new_name)
            out_f.write(line)
            print(line)
            
### тут укореняла в seaview
```

```python
from Bio import SeqIO

file = SeqIO.parse('raw_data/docx_and_txt/seqdump COI others.txt', 'fasta')

with open('trees/ml_tree_6_more_outgroup/RAxML_for_seaview_rooted.newick', 'r') as in_f:
    with open('trees/ml_tree_6_more_outgroup/RAxML_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            
            for rec in file:
                if rec.id in line:
                    new_id = rec.description.replace(' cytochrome oxidase subunit I (COI) gene, partial cds; mitochondrial', '')
                    new_id = new_id.replace(' cytochrome oxidase subunit I (cox1) gene, partial cds; mitochondrial', '')
                    new_id = new_id.replace(' cytochrome oxidase subunit I gene, partial cds; mitochondrial', '')
                    new_id = '"' + new_id + '"'
                    line = line.replace(rec.id, new_id)
                    
            for ind, name in zip(df.index, df.description):
                # print(ind, name)
                name = name.lstrip(' ')
                name = name.replace(',', '')
                if name in line:
                    new_name = '"' + df['description'][ind] + '"'
                    line = line.replace(name, new_name)
                else:
                    print(name)
            out_f.write(line)
            print(line)
```
