```python
with open('/home/yulia/parampampam/trees/ITS/tree_4/RAxML_rooted.newick', 'r') as in_f:
    with open('/home/yulia/parampampam/trees/ITS/tree_4/RAxML_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for ind, name in zip(df.index, df.description):
                # print(ind, name)
                name = name.lstrip(' ')
                name = name.replace(',', '').replace(';', '').replace(':', '').replace('(', '').replace(')', '')
                if name in line:
                    new_name = '"' + df['description'][ind] + '"'
                    line = line.replace(name, new_name)
                else:
                    print(name)
            out_f.write(line)
            print(line)
```

```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle

t = Tree('/home/yulia/parampampam/trees/ITS/tree_4/RAxML_ete_tree.newick', quoted_node_names=True)
# print(t)

for node in t.traverse():
    node.img_style['size'] = 0
    
for n in t.get_leaves():
    n.name = n.name[:70]
    
t.ladderize(direction=1)
ts = TreeStyle()
ts.show_branch_support = True
ts.scale =  1500
ts.branch_vertical_margin = 10
# ts.mode = "c"
t.show(tree_style=ts)
t.render('/home/yulia/parampampam/figures/ITS/new_more/RAxML_ete_tree.pdf', dpi=1200, tree_style=ts)
```
