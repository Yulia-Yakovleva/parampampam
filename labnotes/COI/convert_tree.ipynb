```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle
import pandas as pd

df = pd.read_csv('./raw_data/stats/stats_final.tsv', sep='\t', index_col=0)
# print(df)

with open('trees/ml_tree_3/PhyML_Newick_tree.nhx', 'r') as in_f:
    with open('trees/ml_tree_3/PhyML_Newick_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = df['description'][el].replace(',', '')
                    line = line.replace(el, new_name)
            out_f.write(line)
            print(line)
            
### тут укореняла в seaview
```

```python
with open('trees/ml_tree_3/PhyML_Newick_tree_rooted_100.newick', 'r') as in_f:
    with open('trees/ml_tree_3/PhyML_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for ind, name in zip(df.index, df.description):
                # print(ind, name)
                name = name.lstrip(' ')
                name = name.replace(',', '')
                if name in line:
                    new_name = '"' + df['description'][ind] + '"'
                    line = line.replace(name, new_name)
            out_f.write(line)
            print(line)
```

```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle

t = Tree('trees/ml_tree_3/PhyML_ete_tree.newick', quoted_node_names=True)
# print(t)

for node in t.traverse():
    node.img_style['size'] = 0
    
t.ladderize(direction=1)
ts = TreeStyle()
ts.show_branch_support = True
ts.scale =  1000
ts.branch_vertical_margin = 10
# ts.mode = "c"
# t.show(tree_style=ts)
t.render('trees/ml_tree_3/PhyML_ete_tree_1000.pdf', dpi=1200, tree_style=ts)
```
