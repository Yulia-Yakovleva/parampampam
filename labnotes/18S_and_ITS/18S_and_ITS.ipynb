```python
# для Максима
import pandas as pd

def rename_ids(stats, in_f, out_f):
    df = pd.read_csv(stats, sep='\t')
    with open(in_f) as inf, open(out_f, 'w') as outf:
        for line in inf:
            for idx, descr in zip(df['Unnamed: 0'], df['description']):
                if idx in line:
                    line = line.replace(idx, descr)
            outf.write(line)
                
                
stats = '/home/yulia/parampampam/raw_data/stats/COI/stats_final.tsv'
in_f = '/home/yulia/parampampam/alignments/COI/tree_4_more_outgroup/mafft_aln_fianl_more_outgroup_full.fa'
out_f = '/home/yulia/parampampam/alignments/COI/tree_4_more_outgroup/COI_normal_names.fa'

stats = '/home/yulia/parampampam/raw_data/stats/18S/total_stats_full.tsv'
in_f = '/home/yulia/parampampam/alignments/18S/18S_GBlocks_regions_3.fst'
out_f = '/home/yulia/parampampam/alignments/18S/SSU_normal_names.fa'

stats = '/home/yulia/parampampam/raw_data/stats/ITS/total_stats_full.tsv'
in_f = '/home/yulia/parampampam/alignments/ITS/ITS_GBlocks_2.fst'
out_f = '/home/yulia/parampampam/alignments/ITS/ITS_normal_names.fa'

rename_ids(stats, in_f, out_f)
```
