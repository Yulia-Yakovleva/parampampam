```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle
import pandas as pd

df = pd.read_csv('../../raw_data/stats/18S/total_stats.tsv', sep='\t', index_col=0)
# print(df)

with open('../../trees/18S/tree_2/RAxML_bipartitions.result', 'r') as in_f:
    with open('../../trees/18S/tree_2/RAxML_for_seaview.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for el in df.index:
                if el in line:
                    new_name = df['description'][el].replace(',', '').replace(';', '').replace(':', '').replace('(', '').replace(')', '')
                    line = line.replace(el, new_name)
            out_f.write(line)
            print(line)
            
### тут укореняла в seaview
```

```python
with open('../../trees/18S/tree_2/RAxML_rooted.newick', 'r') as in_f:
    with open('../../trees/18S/tree_2/RAxML_ete_tree.newick', 'w') as out_f:
        for line in in_f:
            line = line.strip()
            for ind, name in zip(df.index, df.description):
                # print(ind, name)
                name = name.lstrip(' ')
                name = name.replace(',', '').replace(';', '').replace(':', '').replace('(', '').replace(')', '')
                if name in line:
                    new_name = '"' + df['description'][ind] + '"'
                    line = line.replace(name, new_name)
                else:
                    print(name)
            out_f.write(line)
            print(line)
```

```python
from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle

t = Tree('../../trees/18S/tree_2/RAxML_ete_tree.newick', quoted_node_names=True)
# print(t)

for node in t.traverse():
    node.img_style['size'] = 0
    
for n in t.get_leaves():
    n.name = n.name[:70]
    
t.ladderize(direction=1)
ts = TreeStyle()
ts.show_branch_support = True
ts.scale =  1500
ts.branch_vertical_margin = 10
# ts.mode = "c"
t.show(tree_style=ts)
t.render('../../trees/18S/tree_2/RAxML_ete_tree_strict.pdf', dpi=1200, tree_style=ts)
```
